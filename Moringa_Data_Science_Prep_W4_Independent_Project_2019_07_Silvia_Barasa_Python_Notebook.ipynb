#1.5 Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.

```python
#Create a new dataframe where city is Paris
df_paris = df[df.city == 'Paris']
df_paris.head()
```

```python
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018
df_paris[df['bluecar_counter'] == 0].groupby('hour')['hour'].value_counts().sort_values(ascending= False).head(1)
```

#1.6 What is the most popular hour for returning cars?

```python
#What is the most popular hour for returning cars?
#we will confirm the hour with the highest number of slots and that will be the best time 
df.groupby('slots')['hour'].value_counts().sort_values(ascending=True).head(10)
```

#1.7.What station is the most popular? Overall?

```python
#What station is the most popular? Overall?
df.groupby('station_type')['public_name'].value_counts().sort_values(ascending=False).head(1)
```

#1.8 What station is the most popular at the most popular picking hour?

```python
#Finding the most popular hour for picking up blue cars
df[df['bluecar_counter'] == 0]['hour'].value_counts().sort_values(ascending=False).head(1)
```

```python
#What station is the most popular at the most popular picking hour?

df[(df['kind'] == 'STATION') & (df['hour'] == 14)].groupby('public_name')['public_name'].count().sort_values(ascending= False).head(1)
```

#1.9 What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code?Overall?At the most popular picking hour?

```python
#what  postal code is the most popular for picking up Blue cars?
df[(df['kind'] == 'STATION') & (df['bluecar_counter'] == 0)].groupby('postal_code')['postal_code'].value_counts().sort_values(ascending= False).head(1)
```

```python
#Does the most popular station belong to that postal code?Overall?
#df[(df['kind'] == 'STATION') & (df['bluecar_counter'] == 0)].
df.loc[(df['postal_code'] == 75008) & (df['public_name'] == 'Paris/Jean-Jacques Rousseau/62')]
#hence it does not belong to that postal code
```

```python
#Does the most popular station belong to that postal code. At the most popular picking hour?
df.loc[(df['kind'] == 'STATION') & (df['postal_code'] == 75008) & (df['hour'] == 14)]
```

#1.10 Do the results change if you consider Utilib and Utilib 1.4 instead of Blue cars? 

###1.10.1 considering utilib

```python
#Finding out the most popular postal code for picking up utilib 
df.loc[df['utilib_counter'] == 0]['postal_code'].value_counts().sort_values(ascending = False).head(1)
```

```python
#Does the most popular station belong to that postal code. At the most popular picking hour?
df.loc[(df['kind'] == 'STATION') & (df['postal_code'] == 75015) & (df['hour'] == 14)]
```

```python
#Finding out if the most popular postal code belongs to the most popular station
df.loc[(df['kind'] == 'STATION') & (df['postal_code'] == 75015) & (df['hour'] == 14) & (df['public_name'] == 'Paris/Jean-Jacques Rousseau/62')]
#hence it doesnot belong to the most popular station
```
